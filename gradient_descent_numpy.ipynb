```python
import numpy as np
from sklearn.datasets import load_boston     #housing price dataset
```

```python
data = load_boston()                        #load dataset
x = data['data'][:,0]                       #slice the first column of dataset
y = data['target']                          #target label
```

```python
def calc_error(x,y,m,b):                    #error calculating function
    error = 0                               #init error as zero
    for i in range(506):                    #iterate over the entire dataset
        x1 = x[i]               
        y1 = y[i]
        error += (y1 - (m * x1 + b))** 2    #MSE(Mean Squared Error)
    total_error = error/506                 #mean of error
    print('Slope = {0}    Y-Intercept = {1}    Error ={2}'.format(round(m,10),round(b,10),round(total_error,10))) #print slope,y-intercept,error
```

```python
def grad_descent(x,y,l_rate,epochs):           #optimization algorith - Gradient Descent
    new_m = 0                                  #init the slope as zero
    new_b = 0                                  #init y-intercept as zero
    b_grad = 0                                 #init gradient(i.e,partial derivative) of y-intercept as zero
    m_grad = 0                                 #init gradient(i.e,partial derivative) of slope as zero
    for i in range(epochs):                    #iterate to reduce error
        m = new_m    
        b = new_b
        for i in range(506):                  #iterate through dataset
            x1 = np.array(x[i])
            y1 = np.array(y[i])
            b_grad += (-2 * (y1 - (m * x1 + b)))        #gradient of y-intercept
            m_grad += (-2 *x1 * (y1 - (m * x1 + b)))    #gradient of slope
        new_m = m - (l_rate * m_grad/506)               #updated slope value
        new_b = b - (l_rate * b_grad/506)               #updated y-intercept value
        calc_error(x,y,new_m,new_b)                     #function call to calculate error value
    
```

```python
#optimum epoch = 129

learning_rate = 0.0001
epochs = 130


grad_descent(x,y,learning_rate,epochs)
```
